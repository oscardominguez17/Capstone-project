## Capstone Project Fuel consumption

#### Description
The idea is to create a model that predicts what the fuel consumption of a car would be at the end of short or medium period, based on previous consumption. for this I am going to use time series to me make a forecast.

#### Business Understanding
Some companies use cards called "fuelcard" that they give to their drivers so they can tank their vehicles and in this way keep track of how much money they spend on gasoline. This is a Dataset that contains information of one vehicle and their consumption during 2017 to 2024.

#### Dataset

The dataset has 622 entries and the next features:
 * date: date on which the car was filled with gasoline
 * unit_id: Vehicle id
 * make: Vehicle brand
 * model: Vehicle model
 * year: year of the vehicle
 * engine: cylinder
 * mpghighway: how many miles you travel per gallon on the highway
 * mpgcity: how many miles you get per gallon in the city
 * tankSizeGallons: tank size
 * status: If the fuelcard is active
 * site_city: city
 * quantity: Tanked quantity
 * unit_of_measure: Unit of measurement in which the vehicle was tanked
 * price: purchase price
 * odometer: odometer consumed since the last purchase
 * currency: currency in which the payment was made
 * site_state: state code

```python
# Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.forecasting.stl import STLForecast
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from pmdarima import auto_arima

import numpy as np
from scipy.stats import norm
```

### Data Understanding

```python
df = pd.read_csv("./dataFuel.csv")
df.info()
```

```python
df.head()
```

```python
## Detecting null values
df.isnull().sum()
```

```python
# This is the main value on which we are going to focus our analysis
value_to_analize = "quantity"
```

## Data Cleaning

```python
# convert the string "date" to datetime format
df["datetime"] = pd.to_datetime(df["date"], errors='coerce')
```

```python
df = df[[value_to_analize, "datetime"]]
print(len(df))
# We have some invalid dates
df.isnull().sum()
```

```python
## Delete NaT in datetime since we need to order the data
df.dropna(inplace=True)
print(len(df))
```

```python
# Set the date time as index
df.set_index(["datetime"], inplace=True)
df
```

### Detect and remove outliers

```python
fig = px.box(y=df[value_to_analize])

# Update the layout
fig.update_layout(
    title="Outliers",
    xaxis_title="Data",
    yaxis_title="Values"
)

# Show the plot
fig.show()
```

```python
# Inter quartil range
q1= df["quantity"].quantile(0.25)
q3= df["quantity"].quantile(0.75)
IQR = q3 - q1
lower_fence = q1 - (1.5 * IQR)
upper_fence = q3 + (1.5 * IQR)
print("Lower Fence: " +  str(lower_fence) + " - " + str(upper_fence))
```

```python
# we eliminate the values below the lower fence and above the upper fence
df_cleaned = df[(df[value_to_analize] >= lower_fence) & (df[value_to_analize] < upper_fence)]
print(len(df_cleaned))
# we deleted 12 rows in this process
df_cleaned
```

Let's visualize the information

```python
df_cleaned.plot()
plt.title("Historical data")
```

#### Initial Conclusions
1) As we can see we have 3 main periods and we also have some years with empty data.
2) The information is not divided by date, there are some days when we have two or more rows and other days without data, since the vehicle is not filled with gasoline every day.

To solve problem 1, we are going to work with data from 2023 onwards, since it is the most recent data, consumption patterns may have changed since 2018, so we are going to work with the most recent data

To solve problem 2, we are going to complete the information on the missing days, dividing the last consumption into the days for which there are no records.

```python
# 1) Select data from 2023 onwards
df_cleaned = df_cleaned[df_cleaned.index > "2023-01-01"]
# tenemos un total de 187 registros
print(len(df_cleaned))
df_cleaned
```

```python
df_hist = df_cleaned[df_cleaned.index < "2024-03-15 00:00:00"]

df_future = df_cleaned[(df_cleaned.index > pd.to_datetime("2024-03-14 23:59:59"))]
print(len(df_future))
df_future.head(50)
```

### Fill missing dates and values

```python
def fill_and_consolidate(df):
    df.index = pd.to_datetime(df.index)
    
    # Extract the date without the time
    df['date'] = df.index.date
    
    # Group by date and sum the values to analyze
    df = df.groupby('date')[value_to_analize].sum()
    df.index = pd.to_datetime(df.index)
    
    # Create a date range covering from the minimum date to the maximum date
    all_dates = pd.date_range(start=df.index.min(), end=df.index.max(), freq='D')
    df = df.reindex(all_dates, fill_value=0)
    
    return df

# Now we have 31 days for the future and 247 for the historical data but with some 0 values we are going to solve this soon
df_hist = fill_and_consolidate(df_hist)
print(len(df_hist))
print(df_hist.head(5))

df_future = fill_and_consolidate(df_future)

print(len(df_future))
```

```python
# Divide the gasoline from the last day the vehicle was refueled by the days with a value of 0
def fill_cero_days(df_with_all_days):
    result_df = df_with_all_days.copy().reset_index()
    before= ''
    for index, item in result_df.iterrows():
        if(index == len(result_df)-1):
            break;
        else:
            next_item = result_df.iloc[index + 1]
            if(next_item[value_to_analize] == 0):
                zero_indexs = []
                for y in range(index+1, len(df_with_all_days)):
                    if(result_df.iloc[y][value_to_analize] == 0):
                        zero_indexs.append(y)
                    else:
                        break;      
                zero_values = len(zero_indexs)
                new_price = item[value_to_analize] / (zero_values +1)
                result_df.loc[index, value_to_analize] = new_price
                for ind in zero_indexs:
                    result_df.loc[ind, value_to_analize] = new_price
    return result_df;
                

df_hist = fill_cero_days(df_hist)

df_future = fill_cero_days(df_future)
df_future.head()
```

Now that we had all the days with a value let's plot the info

```python
df_hist.rename({"index": "date"},axis=1, inplace=True)
df_hist.set_index("date", inplace=True)

df_future.rename({"index": "date"},axis=1, inplace=True)
df_future.set_index("date", inplace=True)
```

```python
plt.figure(figsize=(10,8))
plt.plot(df_hist)
plt.title("Consolidated historical data")
plt.ylabel("Quantity")
plt.xlabel("Dates")
plt.show()
```

### Autocorrelation (ACF) and Partial Autocorrelation (PACF) Historical Data

```python
fig, (ax1, ax2) = plt.subplots(2, 1,figsize=(12,8))
plot_acf(df_hist[value_to_analize], lags=30, ax= ax1)
plot_pacf(df_hist[value_to_analize], lags=30, ax= ax2)
plt.title("Sample Autocorrelation Historical Data")
plt.show()
```

### Evaluating stationarity

```python
decomposition = seasonal_decompose(df_hist, model='additive')

# Graficar las componentes
plt.figure(figsize=(10, 6))
plt.subplot(411)
plt.plot(df_hist, label='Original')
plt.title("Original Data Fueldcard")
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color="green")
plt.title("Trend Data Fueldcard")
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality', color="yellow")
plt.title("Seasonality Data Fueldcard")
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(decomposition.resid, label='Residue', color="red")
plt.title("Residue Data Fueldcard")
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

Trend: We have not a clear trend in the data.csv

Seasonality: We have certain peaks each time the vehicle's tank is filled and we have a drop in the days when it is not necessary to fill the vehicle's tank.

```python
# function to evaluate if a time serie is stationary or not, 
# it receivies by parameter the Argumented Dickey-fuller test.
def isStationary(adf_data):
    if adf_data[0] < adf_data[4]['5%']:
        return "yes"
    else:
        return "no"
```

```python
# We can reject the null hypothesis.
adfuller_hist = adfuller(df_hist)
isStationary(adfuller_hist)
```

### Arima Model

For this analysis I started with the parameters 5,0,5 since after analyzing the ACF and PACF plots
It looks like these have some autocorrelation to cover at lag 5.

```python
# Evaluating the ACF, MA("q") parameter 5
q = 5
# Evaluating the PACF for AR("p") parameter 5
p = 5

d = 0

arma5 = ARIMA(df_hist, order = (p, d, q)).fit()
arma5
```

```python
arma5.summary()
# AIC with parameters (5,0,5) = 1801
```

To help me find the best parameters I used auto_arima which helps me with this objective through Grid Search.

```python
hyper_params = auto_arima(df_hist[value_to_analize], trace= True)
hyper_params.summary()
```

According to this, the best parameters for our model were 1,1,1, let's check it.

```python
# Evaluating the ACF, MA("q") parameter 1
q = 1
# Evaluating the PACF for AR("p") parameter 1
p = 1

d = 1

arma = ARIMA(df_hist, order = (p, d, q)).fit()
arma.summary()
```

* parameters 5,0,5 = AIC 1801.350
* parameters 1,1,1 = AIC 1795.340

The Akaike Information Criterion (AIC) with parameters 5,0,5 is greater than 1,1,1, so we can conclude that 1,1,1 is less complex.

#### Making predictions

```python
hist_preds = ''

hist_preds = arma.predict()

plt.figure(figsize = (12, 4))
plt.plot(hist_preds, label = 'model')
plt.plot(df_hist, label = 'data')
plt.legend()
plt.grid()
plt.title('Result of ARMA Model');
```

### Model Residual

```python
# residual autocorrelation
# more than one means the model needs to be adjusted
%matplotlib inline
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = plot_acf(arma.resid, lags=31, ax=ax1)
plt.title("Residual Autocorrelation")
```

As we can see in the previous plot, at lag 5 there still seems to be autocorrelation.
but we avoid processing the model with q = 5 to avoid overfitting, it is still ok.

```python
# Custom function to test for a normal distribution in the residual data
def resid_histogram(data):
    import matplotlib.pyplot as plt
    from numpy import linspace
    from scipy.stats import norm

    plt.figure(figsize=(10,6))
    plt.hist(data, bins = 'auto', density = True, rwidth = 0.85,
             label = 'Residuals')
    mean_resid, std_resid = norm.fit(data)
    xmin, xmax = plt.xlim()
    curve_length = linspace(xmin, xmax, 100)
    bell_curve = norm.pdf(curve_length, mean_resid, std_resid)
    plt.plot(curve_length, bell_curve, 'm', linewidth = 2)
    plt.grid(axis='y', alpha = 0.2)
    plt.xlabel('Residuals')
    plt.ylabel('Density')
    plt.title('Residuals vs Normal Distribution - Mean = '+str(round(mean_resid,2))+', Std = '+str(round(std_resid,2)))
    plt.show()
```

Let's review the residues for the arma model

```python
# it Looks near to a normal distribution, there is only a peak around -9 which is a lot higher than the curve
resid_histogram(arma.resid)
```

### Forecast last month

```python
stlf = STLForecast(df_hist[value_to_analize], ARIMA, model_kwargs={'order':(1, 1, 1)})
#fit model using historical data
stlf_results = stlf.fit()
#produce forecast for future data
forecast = stlf_results.forecast(len(df_future))
```

```python
plt.plot(df_future[value_to_analize], label = 'true future data')
plt.plot(forecast, label = 'forecast')
plt.plot(df_hist[value_to_analize], label = 'training data')
plt.legend()
plt.title('Forecast with STL and Future Data Stationary')
plt.grid();
plt.xticks(rotation = 90)
plt.show()
```

```python
total_monthly_forecast = forecast.sum()
total_real_future = df_future.sum()
print("Last Month forecast: "+str(total_monthly_forecast)+" VS Total real future: "+ str(total_real_future))
```

```python
pred_error = df_future[value_to_analize] - forecast
mae = np.mean(abs(pred_error))
rmse = np.sqrt(np.mean(pred_error ** 2))

print(f'MAE: {mae:.2f}')
print(f'RMSE: {rmse:.2f}')

# 7.650244581241624
# RMSE: 9.129403669077067
```
